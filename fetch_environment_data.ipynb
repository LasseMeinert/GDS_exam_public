```python
from utilities import *
```

```python
data = pd.read_csv('data/final_dataset.csv',sep=';')
```

```python
urls = data.final_url.tolist()
```

```python
%%time
air, noise, energy, faulty = [],[],[],[]

url_dict = {}

for idx,url in enumerate(urls):
    if idx % 500 == 0:
        print(idx)
    
    curr_url = {}
    
    try:
        a,n,e = get_pollution(url)
        
        curr_url['air'] = a
        curr_url['noise'] = n
        curr_url['energy'] = e
        curr_url['faulty'] = False


    except:
        
        curr_url['air'] = [None,None,None,None]
        curr_url['noise'] = 'Unknown'
        curr_url['energy'] = 'Unknown'
        curr_url['faulty'] = True
        print(url)
    
    curr_url['idx'] = idx
    url_dict[url] = curr_url
```

```python
meta_df = pd.DataFrame.from_dict(url_dict, orient='index')
```

```python
meta_df = meta_df.reset_index()
meta_df.columns = ['final_url','air','noise','energy','faulty','idx']
```

```python
data = data.merge(meta_df,on='final_url')
```

```python
data.to_csv('data/metadata.csv',sep=';')
```
